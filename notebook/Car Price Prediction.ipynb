#### Checking for overfitting and data mismatch


To check if the model is either overfitting or there is a data mismatch, a train-dev set which is held out from the original training data is used. If the model performs well on the train-dev set, it means it does not overfit.
After checking the performance of the model on the validation or dev set, you then compare it with its performance on the train-dev set. If the difference in performance is stark, then there is a data mismatch.

```python
# Helper function for checking overfitting
# and data mismatch
def calculate_dev_set_scores(model):
    print(f'Train score for {model}: {model.score(X_train, y_train)}')
    print(f'Train-dev score {model}: {model.score(X_train_dev, y_train_dev)}')
    print(f'Validation score {model}: {model.score(X_val, y_val)}')
    print('\n')
```

```python
[calculate_dev_set_scores(model) for model in [lr, svm, rf]]
```

It can then be concluded that no model is overfitting the data, neither is there a data mismatch.

```python
class AutoVivification(dict): # dynamic dictionary
    def __getitem__(self, item):
        try:
            return dict.__getitem__(self, item)
        except KeyError:
            value = self[item] = type(self)()
            return value
```

```python
def evaluate_models(models, scorers, X_test, y_test):
    """Evaluate trained models on given scores and prints out the metric score of each trained model.
    models:
        Must be a list of tuples with each tuple having the model name and the trained model instance.
    scorers:
        Must be a list of tuples with each tuple having the score name and the scorer instance.
    X_test:
        The feature test set. Must be of size (n_features, n_samples)
    y_test:
        The target test set. Must be of size (n_samples,)
     """
    model_scores = AutoVivification()
    for(model_name, model) in models:
        print(f'Test Scores of {model_name} ')
        for (score_name, scorer) in scorers:        
            test_score = scorer(y_true=y_test, y_pred=model.predict(X_test))
            model_scores[model_name][score_name] = test_score
            print(f'{score_name} : {test_score:.4f}')
        print('\n')
```

```python
from sklearn.metrics import r2_score

def find_mape(y_true, y_pred):

    if not all([isinstance(y_pred, np.ndarray),
                isinstance(y_true, np.ndarray)]):
        y_pred, y_true = np.array(y_pred),\
        np.array(y_true)
        
    percentage_error = np.absolute((y_pred - y_true)/ y_true) * 100
    mape = round(np.mean(percentage_error), 2)
    return mape
```

```python
mape = find_mape
scorers = [('r2', r2_score), ('mape', mape)]
models = [('Linear Regression', lr), ('SVM', svm), ('Random Forest', rf)]

evaluate_models(models=models, scorers=scorers, X_test=X_test, y_test=y_test)
```
